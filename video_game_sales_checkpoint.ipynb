```python
df1 = data[["Year_of_Release", "NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]
#supprimer les repititions 
df1=df1.drop_duplicates(['Year_of_Release'])
df1=df1.sort_values(by='Year_of_Release', ascending=True)
df1[df1.duplicated(subset=['Year_of_Release'], keep='last')]
df1
```

```python
#####
plt.subplots(figsize=(15,10))
#####
y=pd.Series(df1['Year_of_Release'])
x1=pd.Series(df1['NA_Sales'])
x2=pd.Series(df1['EU_Sales'])
x3=pd.Series(df1['JP_Sales'])
x4=pd.Series(df1['Other_Sales'])
x5=pd.Series(df1['Global_Sales'])

#####
plt.plot(y, x1, label='NA_Sales')
plt.plot(y, x2, label='EU_Sales')
plt.plot(y, x3, label='JP_Sales')
plt.plot(y, x4, label='Other_Sales')
plt.plot(y, x5, label='Global_Sales')
#####
plt.xlabel('Year_of_rel')
plt.ylabel('nbr_sales (In millions)')
plt.title('SALES')
plt.legend()
plt.show()
```

```python
df2=data[["Genre", "JP_Sales"]]
df2_gb = df2.groupby('Genre')
df2_gb=df2_gb.sum()
df2_gb['%'] = (df2_gb['JP_Sales']/df2_gb['JP_Sales'].sum())*100
df2_gb
```

```python
plt.figure(figsize=(13,10))
df2_gb['JP_Sales'].plot(kind='pie',subplots=True)
plt.title('Japon')
```

```python
df2=data[["Genre", "Global_Sales"]]
df2_gb = df2.groupby('Genre')
df2_gb=df2_gb.sum()
df2_gb['%'] = (df2_gb['Global_Sales']/df2_gb['Global_Sales'].sum())*100
print(df2_gb)
plt.figure(figsize=(13,10))
df2_gb['Global_Sales'].plot(kind='pie',subplots=True)
plt.title('ALL THE WORLD')
```
